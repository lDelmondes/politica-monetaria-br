## Objetivo

Verificar que o ambiente Python isolado (`.venv`) está corretamente configurado e que 
as bibliotecas essenciais para análise econométrica de séries temporais estão instaladas 
e acessíveis pelo notebook.

## Contexto do projeto

Este notebook é o primeiro de uma série que constrói uma análise dos efeitos da política 
monetária sobre PIB e inflação no Brasil, no período 2003-2025, em frequência mensal, 
utilizando modelos VAR e BVAR. O trabalho replica e estende a metodologia da monografia 
de graduação do autor, com três melhorias principais:

1. **Frequência mensal** em vez de trimestral, utilizando IBC-Br como proxy do PIB
2. **Inclusão da expectativa de inflação do Focus** como controle para o *price puzzle*
3. **Tratamento explícito da pandemia** via variáveis dummy

## Stack técnica

- **pandas**: manipulação de DataFrames e séries temporais
- **numpy**: operações numéricas e álgebra linear
- **matplotlib**: visualização
- **statsmodels**: testes de estacionariedade, VAR, VECM, X-13 ARIMA-SEATS
- **python-bcb**: interface para APIs do Banco Central do Brasil
